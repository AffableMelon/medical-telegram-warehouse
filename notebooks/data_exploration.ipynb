# Data Exploration

This notebook demonstrates how to load and explore the scraped telegram data.

```python
import json
import glob
import pandas as pd
import os

# Set path to data
data_path = "../data/raw/telegram_messages/*/*.json"

# Load all JSON files
all_messages = []
for f in glob.glob(data_path):
    with open(f, 'r') as infile:
        data = json.load(infile)
        all_messages.extend(data)

# Create DataFrame
df = pd.DataFrame(all_messages)
print(f"Total messages loaded: {len(df)}")
df.head()
```
